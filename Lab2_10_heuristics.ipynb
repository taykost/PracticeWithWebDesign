### Match between system and the real world
The system should speak the users' language, with words, phrases and concepts familiar to the user, rather than system-oriented terms. Follow real-world conventions, making information appear in a natural and logical order.




Screenshot 2020-09-16 140731.png

There are icons in the top right corner with words under them and also icons without any words. The icons with the words are pictures that represent what the words mean and make the words more understandable. It would make sense that the "my information" has a person icon since this information relates to the user. It would also make sense that the "my finances" has a dollar sign icon since this information relates to money. Lastly, it would make sense that the "my advisors" has an icon with two people because this shows the relationship between the user and their advisor. The icons that don't have any words are self explanatory and easy to understand what they would do. The house icon lets the user know that when clicked, the user will be taken back to the home page. The magnifying glass icon lets the user know that when clicked, the user will be able to search for something within the website. The gear icon lets the user know that when clicked, the website settings will come up. All of these icons are universal and are easy to understand because their purpose in real life corresponds closely with the actions being performed within the website.

Users often choose system functions by mistake and will need a clearly marked "emergency exit" to leave the unwanted state without having to go through an extended dialogue. Support undo and redo.

### Consistency and standards
Users should not have to wonder whether different words, situations, or actions mean the same thing. Follow platform conventions.



Screenshot 2020-09-16 155619.png

On the website, each page excluding the home page, has the same layout. After clicking on a page and being redirected, the new page will have a side navigation on the left with different options. The current option selected will then be shown in the large viewing pane to the right of the navigation. This is the layout for each page, which is consistent and convienent. The user will know what to expect when switching pages.
